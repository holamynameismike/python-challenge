```python
import csv

# Read the CSV file and initialize variables
total_votes = 0
candidate_votes = {}
winner = ""
max_votes = 0

# Open the CSV file
with open("election_data.csv") as file:
    reader = csv.reader(file)

    # Skip the header row
    next(reader)

    # Process each row in the CSV file
    for row in reader:
        # Increment the total vote count
        total_votes += 1

        # Get the candidate's name from the row
        candidate = row[2]

        # Add the candidate to the dictionary if not already present
        if candidate not in candidate_votes:
            candidate_votes[candidate] = 0

        # Increment the candidate's vote count
        candidate_votes[candidate] += 1

        # Check if the current candidate has more votes than the previous winner
        if candidate_votes[candidate] > max_votes:
            winner = candidate
            max_votes = candidate_votes[candidate]

# Calculate the percentage of votes for each candidate
candidate_percentages = {
    candidate: (votes / total_votes) * 100
    for candidate, votes in candidate_votes.items()
}

# Format the results as strings
total_votes_str = f"Total Votes: {total_votes}"
candidate_results = [
    f"{candidate}: {candidate_percentages[candidate]:.3f}% ({votes})"
    for candidate, votes in candidate_votes.items()
]
winner_str = f"Winner: {winner}"

# Print the analysis to the terminal
print("Election Results")
print("-------------------------")
print(total_votes_str)
print("-------------------------")
for result in candidate_results:
    print(result)
print("-------------------------")
print(winner_str)
print("-------------------------")

# Export the analysis to a text file
with open("election_results.txt", "w") as file:
    file.write("Election Results\n")
    file.write("-------------------------\n")
    file.write(total_votes_str + "\n")
    file.write("-------------------------\n")
    for result in candidate_results:
        file.write(result + "\n")
    file.write("-------------------------\n")
    file.write(winner_str + "\n")
    file.write("-------------------------\n")

print("Analysis has been exported to 'election_results.txt'.")
```
